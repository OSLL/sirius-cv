```
# Importing modules
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

# Reading the image
img = cv.imread('/content/photo.jpg')

# Height and width, shifts of height and width of the image
y, x = img.shape[:2]
sy, sx = y // 4, x // 4
# Slicing the image with our parameters
edit_img = img[sy : y-sy, sx : x-sx]

# Horizontal flipping the image
edit_img = cv.flip(edit_img, 1)

# Showtime!
cv2_imshow(edit_img)
cv.waitKey(0)
```
