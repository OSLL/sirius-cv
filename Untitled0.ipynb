```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
!wget  https://github.com/OSLL/sirius-cv/raw/12_Daniil_Komarov_object_detection/to_find.jpg
img1 = cv2.imread('/content/to_find.jpg', cv2.IMREAD_GRAYSCALE)

!wget https://github.com/OSLL/sirius-cv/raw/12_Daniil_Komarov_object_detection/picture.jpg
img2 = cv2.imread('/content/picture.jpg', cv2.IMREAD_GRAYSCALE)

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x:x.distance)

result = cv2.drawMatches(img1, kp1, img2, kp2, matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
cv2_imshow(result)
```
